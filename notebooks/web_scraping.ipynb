In this notebook I will get the data I need for the project through web scraping from the website cultbeauty.com. 
I will get the name, price and rating (stars) for a variety of products in different categories: moisturisers, cleansers, toners, masks, serums etc. After I have those, I will get a list of which of tehse products products are in other categories such as vegan, spf, gluten free etc. I will add all this information to the data frame and it will be used as features in the clustering process. 

```python
import requests
from bs4 import BeautifulSoup
```

```python
import pandas as pd
```

```python
def get_data(soup): 
    rating_list = []
    for div in soup.find_all("span", {"class":"productBlock_ratingValue"}):
        rating_list.append(div.get_text("span"))
    
    index_list = []
    
    block_list = soup.find_all("div", {"class":"productBlock"})
    
    for index in range(len(block_list)):
        for child in block_list[index].descendants:
            if 'Stars' in str(child.string):
                    index_list.append(index)

    index_list = [index_list[i*2] for i in range(int(len(index_list)/2))]
    
    df3 = pd.DataFrame({'stars':rating_list}, index=index_list)
    
    prices_list = []

    for div in soup.find_all("div", {"class":"productBlock_priceBlock"}):
        value_list = div.get_text('span').split('span')
        if len(value_list) == 7:
            prices_list.append(value_list[4])
        elif len(value_list) == 12:
            prices_list.append(value_list[4])
        else:
            prices_list.append(value_list[3])
        
    df2 = pd.DataFrame({'price':prices_list}) 
    
    names_list = []
    for div in soup.find_all("div", {"class":"productBlock_title"}):
        names_list.append(div.get_text("h3").split('\n')[2].strip())
    
    df1 = pd.DataFrame({'names':names_list})
    
    moisturisers = pd.concat([df1, df2, df3], axis=1)
    
    return moisturisers
    
```

```python
def clean_name(name):
    if 'ml' in name or 'oz' in name:
        name_list = name.split(' ')
        return ' '.join(name_list[:-1])
    else:
        return name
```
